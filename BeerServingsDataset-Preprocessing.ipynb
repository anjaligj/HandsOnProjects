For filling missing value

```python
for col in ['beer_servings', 'spirit_servings', 'wine_servings',
       'total_litres_of_pure_alcohol']:
  data[col]=data[col].fillna(data[col]).median()
```

```python
data.isna().sum()
```

```python
##Encoding
data.columns
```

```python
#As apart of encoding droping country column 
data.drop('country',axis=1,inplace=True)

```

```python
#Label encoding
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
label_enc=LabelEncoder()
#label encode continent
data['continent']=label_enc.fit_transform(data['continent'])
```

```python
data.head()
```

```python
data.dtypes
```

```python
#Splitting to target and features
#X stores independent variables,features
X=data.drop(['total_litres_of_pure_alcohol'],axis=1)
#Y has only target variable
Y=data['total_litres_of_pure_alcohol']
X
```

```python
Y
```

```python
#Dropping continent
X1=X.drop('continent',axis=1)
```

```python
X1
```

```python
#Normalising X1
X1=preprocessing.normalize(X1)
```

```python
type(X1)
```

```python
col=['beer_servings', 'spirit_servings', 'wine_servings']
```

```python
X1=pd.DataFrame(X1,columns=col)
X1
```

```python
X1.describe()
```

```python
X.drop(['beer_servings', 'spirit_servings', 'wine_servings'],axis=1,inplace=True)
```

```python
X
```

```python
X=pd.concat([X,X1],axis=1)

```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
#building model
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(X_train,Y_train)
predict=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE=',mean_squared_error(Y_test,predict))
```
